```python
f=open('WhatsApp Chat with MMM.txt','r',encoding='utf-8')
```

```python
#reading data of file in string format
data=f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
#regular expression: string to regx
pattern='\d{1,2}/\d{1,2}/\d{2},\s\d{1}:\d{2}\spm\s-\s'
```

```python
messages=re.split(pattern,data)
messages
```

```python
#[1:]  to remove the starting 1 string that is empty itself
messages=re.split(pattern,data)[1:]
messages
```

```python
#now we have dates as separate data
dates=re.findall(pattern,data)
dates
```

```python
for i in range(0,len(dates)):
    dates[i]=dates[i][0:9]+" "+str(int(dates[i][10])+12)+dates[i][11:14]
#         print(str(int(dates[i][10])+12))
    
```

```python
dates
```

```python
#now lets make dataframe
df=pd.DataFrame({'usermessage':messages,'messagedate':dates})
#convert message datatype
# df['messagedate']=pd.to_datetime(df['messagedate'],format='%d/%m/%Y, %H:%M')
df.rename(columns={'messagedate':'date'},inplace=True)
df.head()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df.shape
```

```python
messages
```

```python
#now we need to separate user name and its msg in two different columns
users=[]
messages=[]
for message in df['usermessage']:
    entry=re.split('([\w\W]+?):\s',message)
    if entry[1:]: #user name
        users.append(entry[1])
        messages.append(entry[2])
    else: #if no username then its group notifcation
        users.append('groupnotification')
        messages.append(entry[0])
df['user']=users
df['message']=messages
df.drop(columns=['usermessage'],inplace=True)
df.head()
```

```python
#now extracting date as month year day etc
df['year']=df['date'].dt.year
```

```python
tempdf=df[df['user']!='groupnotification']
temp=tempdf[tempdf['message']!='<Media omitted>\n']
```

```python
#english stop words were easy to remove
#but we need to remove hinglish words like abey,hai,kya

#so lets remove both english and hindi stop words
f=open('stop_hinglish.txt','r')
stopwords=f.read()
stopwords
```

```python
words=[]
for message in temp['message']:
    for word in message.lower().split():
    #checking that it should not be present in stopwords
        if word not in stopwords:
            words.append(word)
   
```

```python
df['dayname']=df['date'].dt.day_name()
```

```python
df['dayname'].value_counts()
```

```python
period=[]

for hour in df[['dayname','hour']]['hour']:
    if hour==23:
        period.append(str(hour)+"-"+str("00"))
    elif hour==0:
        period.append(str("00")+"-"+str(hour+1))
    else:
        period.append(str(hour)+"-"+str(hour+1))
```

```python
df['period']=period
```
